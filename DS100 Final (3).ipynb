```python
import pandas as pd
import altair as alt
from datetime import datetime
from dateutil.parser import parse
import pandas_datareader as DataReader
```

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
df = pd.read_csv("crypto-markets.csv")
df
```

```python
df.dtypes
len(df.name.unique())
```

```python
# visualize the trends in the data
# for the purpose of this project we will only look at the most popular cryptocurrenct assets,
# Etherum, Bitcoin, and XRP.
# get df of monthly data so altair can graph it
df_monthly = df
df_monthly.set_index('date', inplace = True)
df_monthly.index = pd.to_datetime(df_monthly.index)
df_monthly['date'] = df_monthly.index
df_monthly
```

```python
# Put Data into monthly intervals so altair can read it
df_bitcoin = df_monthly[df_monthly['name']=='Bitcoin'].resample('1M').mean()
df_bitcoin['date'] = df_bitcoin.index
df_xrp = df_monthly[df_monthly['name']=='XRP'].resample('1M').mean()
df_xrp['date'] = df_xrp.index
df_eth = df_monthly[df_monthly['name']=='Ethereum'].resample('1M').mean()
df_eth['date'] = df_eth.index
df_bitcoin
```

```python
df_bitcoin[['close', 'volume','market']].describe()
```

```python
# visualize the data

bit = alt.Chart(df_bitcoin).mark_line().encode(
    x = alt.X('date:T', title='Year'),
    y = alt.Y('close', axis=alt.Axis(title='Closing Price in USD ($)')),
    color=alt.value('orange'),
).properties(
    title='Bitcoin Historical Price'
).configure_title(
    fontSize=20,
    font='Courier',
    anchor='start',
    color='gray'
)

eth = alt.Chart(df_eth).mark_line().encode(
    x = alt.X('date:T', title='Month/Year'),
    y = alt.Y('close', axis=alt.Axis(title='Closing Price in USD ($)')),
    color = alt.value('purple'),
)

xrp = alt.Chart(df_xrp).mark_line().encode(
    x = alt.X('date:T', title='Year'),
    y = alt.Y('close', axis=alt.Axis(title='Closing Price in USD ($)')),
    color = alt.value('navy')
)

bit
```

```python
eth.configure_title(
    fontSize=20,
    font='Courier',
    anchor='start',
    color='gray'
).properties(
    title = "Etherum Historical Prices"
)
```

```python
# XRP follows a very similar pattern as Bitcoin and Ethereum but on a smaller scale
# because of its larger supply.
xrp.properties(
    title = "XRP Historical Prices"
).configure_title(
    fontSize=20,
    font='Courier',
    anchor='start',
    color='gray')
```

```python
# Correltion between S&P 500
# 2013-04-28 to 
start = datetime.strptime("2013-04-28", '%Y-%m-%d')
end = datetime.strptime("2018-11-30", '%Y-%m-%d')

SP500 = DataReader.DataReader(['sp500'], 'fred', start, end)
df_bit = df_monthly[df_monthly['name']=='Bitcoin']['close']
df_bit
SP500['Bitcoin'] = df_bit
SP500.dropna()
graph = SP500.resample('1M').mean()
graph['date'] = graph.index
```

```python
# graph the movement of the two instruments
sandp = alt.Chart(graph).mark_line().encode(
    x='date:T',
    y = alt.Y('sp500', axis=alt.Axis(title='Closing Price in USD ($)')),
    color=alt.value('green'),
)

bit2 = alt.Chart(graph).mark_line().encode(
    x='date:T',
    y = alt.Y('Bitcoin', axis=alt.Axis(title='Closing Price in USD ($)')),
    color=alt.value('orange'),
)

(sandp + bit2).configure_title(
    fontSize=20,
    font='Courier',
    anchor='start',
    color='gray'
).properties(
    title = "S&P 500, Bitcoin Returns"
)
```

```python
# Correlation Between bitcoin, s&p 500
SP500.corr()
```

```python
# visualized well with scatter chart
scatter_plot = alt.Chart(graph).mark_point().encode(
    x='Bitcoin',
    y='sp500',
    color=alt.value('red'),
).mark_circle(size=30)
(scatter_plot + scatter_plot.transform_regression('sp500', 'Bitcoin').mark_line()).configure_title(
    fontSize=20,
    font='Courier',
    anchor='start',
    color='gray'
).properties(
    title = "Correlation Between Bitcoin, S&P 500"
)
```
